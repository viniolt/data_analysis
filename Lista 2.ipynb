## Item a - Usar um pacote pronto do python para aplicar a regressão linear na base de dados. Fazer os gráficos para a base de dados e para a regressão linear

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [6,58,76,58,23,6,75,33,89,41,2,97,3,43,59]
y = [10,12,26,29,31,51,51,53,58,66,67,77,78,89,95]

print("X = ",x)
print("Y = ",y)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, color='skyblue')
plt.show()
```

```python
b, a, _, _, _ = stats.linregress(x, y)
print(f'A equação da reta é: y = {b:.2f}x + {a:.2f}')
```

```python
x = np.array([0,110])
y = a + b*x
plt.figure(figsize=(10,6))
plt.scatter(x, y, color='skyblue')
plt.ylim(10,90)  # Ajuste manual do eixo Y para visualizar melhor a reta
plt.xlim(0, 110)  # Ajuste do eixo X
plt.plot(x, y,'-r', label=f'Regressão Linear y = {b:.2}x + {a:.2}')
plt.show()
```

# Item b - Encontrar os coeficiêntes a e b, usando as fórmulas dadas em aula programando suas próprias funções. Fazer os gráficos para a base de dados e para a regressão linear

```python
def media(x):
    media = 0
    for i in range (len(x)):
        media += x[i]
    return media/len(x)

def regressao(x,y):
    x_mean = media(x)
    y_mean = media(y)

    numerador = 0
    denominador = 0
    for i in range(len(x)):
        numerador += (x[i] - x_mean) * (y[i] - y_mean)
        denominador += (x[i] - x_mean)**2

    b = (numerador/denominador)
    a = y_mean - b*x_mean
    return a, b
```

```python
x = [6,58,76,58,23,6,75,33,89,41,2,97,3,43,59]
y = [10,12,26,29,31,51,51,53,58,66,67,77,78,89,95]
```

```python
a, b = regressao(x,y)
print(f'A equação da reta é: y = {b:.2f}x + {a:.2f}')
```

```python
x1 = 0
x2 = 110
y1 = a + b*x1
y2 = a + b*x2

plt.figure(figsize=(10,6))
plt.scatter(x, y, color='skyblue')
plt.plot([x1,x2], [y1,y2], label=f'Regressão Linear y = {b:.1}x + {a:.4}')
plt.legend()
plt.show()
```

```python
import mlcroissant as mlc
import pandas as pd

# Fetch the Croissant JSON-LD
croissant_dataset = mlc.Dataset('https://www.kaggle.com/datasets/mirichoi0218/insurance/croissant/download')

# Check what record sets are in the dataset
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)

# Fetch the records and put them in a DataFrame
df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))
df.head()
```

```python
x = df['insurance.csv/bmi']
y = df['insurance.csv/charges']
```

```python
plt.figure(figsize=(20,12))
plt.scatter(x, y, color='skyblue')
plt.xlabel('Índice de Massa Corporal')
plt.ylabel('Valor da cobrança hospitalar')
plt.show()
```

```python
b, a, _, _, _ = stats.linregress(x, y)
print(f'A equação da reta é: y = {b:.2f}x + {a:.2f}')
```

```python
x = np.array([0,100])
y = a + b*x

plt.figure(figsize=(10,6))
plt.scatter(x, y, color='skyblue')
plt.plot(x, y,'-r', label=f'Regressão Linear y = {b:.4}x + {a:.5}')
plt.xlabel('Índice de Massa Corporal')
plt.ylabel('Valor da cobrança hospitalar')
plt.legend()
plt.show()
```

## Item b - Encontrar os coeficiêntes a e b, usando as fórmulas dadas em aula programando suas próprias funções. Fazer os gráficos para a base de dados e para a regressão linear.

```python
def media(x):
    media = 0
    for i in range (len(x)):
        media += x[i]
    return media/len(x)

def regressao(x,y):
    x_mean = media(x)
    y_mean = media(y)

    numerador = 0
    denominador = 0
    for i in range(len(x)):
        numerador += (x[i] - x_mean) * (y[i] - y_mean)
        denominador += (x[i] - x_mean)**2

    b = (numerador/denominador)
    a = y_mean - b*x_mean
    return a, b

a, b = regressao(x,y)
print(f'A equação da reta é: y = {b:.2f}x + {a:.2f}')
```
